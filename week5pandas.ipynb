Indexing into specific row(s) and column(s) of the dataframe:

`.loc` is labeled based

```python
df.loc[[0, 1, 2, 3, 4], ['age', 'sex', 'cp']]
```

`iloc` is integer location based

```python
df.iloc[0]
```

You can also slice into the index rows using `iloc`

```python
df.iloc[0:10]
```
